```python
categorical = df[["gender","ssc_b","hsc_b","hsc_s","degree_t","workex","specialisation","status"]]
contineous = df[["ssc_p","hsc_p","degree_p","etest_p","mba_p"]]
```

```python
temp = pd.get_dummies(categorical,drop_first=True)
```

```python
temp.tail()
```

```python
df = pd.concat([contineous,temp],axis=1)
```

```python
df.head()
```

Let us Apply Machine Learning Now

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve, auc
```

```python
X = df.drop(["status_Placed"],axis=1)
y= df["status_Placed"]
```

```python
X_train,X_test,y_train,y_test= tts(X,y,random_state=42,test_size=0.25)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

```python
print(classification_report(y_test, y_pred))
```

Accuracy with Logistic Regression is 90%
